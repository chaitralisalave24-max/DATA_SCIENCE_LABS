Aim

To implement an Unsupervised Generative Model using Autoencoder in TensorFlow and Keras.

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (
    Dense,
    Flatten,
    Reshape,
    Dropout
)

# Load MNIST Handwritten Dataset
(train_images, _), (test_images, _) = mnist.load_data()

# Normalize Images
train_images = train_images.astype("float32") / 255.0
test_images = test_images.astype("float32") / 255.0

# Add Channel Dimension
train_images = np.expand_dims(
    train_images,
    axis=-1
)

test_images = np.expand_dims(
    test_images,
    axis=-1
)

print("Training Dataset Shape :", train_images.shape)
print("Testing Dataset Shape :", test_images.shape)

# Build Autoencoder Network
generative_model = Sequential()

# Encoder Section
generative_model.add(
    Flatten(input_shape=(28,28,1))
)

generative_model.add(Dense(
    256,
    activation='relu'
))

generative_model.add(Dropout(0.2))

generative_model.add(Dense(
    128,
    activation='relu'
))

generative_model.add(Dense(
    32,
    activation='relu'
))

# Decoder Section
generative_model.add(Dense(
    128,
    activation='relu'
))

generative_model.add(Dense(
    256,
    activation='relu'
))

generative_model.add(Dense(
    784,
    activation='sigmoid'
))

# Reshape Output Image
generative_model.add(
    Reshape((28,28,1))
)

# Display Model Summary
generative_model.summary()

# Compile Model
generative_model.compile(
    optimizer='rmsprop',
    loss='mean_squared_error',
    metrics=['accuracy']
)

# Train Generative Model
training_history = generative_model.fit(
    train_images,
    train_images,
    epochs=8,
    batch_size=128,
    validation_split=0.2,
    verbose=1
)

# Evaluate Model
model_loss, model_accuracy = generative_model.evaluate(
    test_images,
    test_images,
    verbose=0
)

print("\nAutoencoder Loss :", model_loss)
print("Autoencoder Accuracy :", model_accuracy)

# Generate Reconstructed Digits
reconstructed_output = generative_model.predict(
    test_images[:6]
)

# Display Original and Generated Images
plt.figure(figsize=(12,4))

for image_index in range(6):

    # Original Image
    plt.subplot(2,6,image_index + 1)

    plt.imshow(
        test_images[image_index].reshape(28,28),
        cmap='binary'
    )

    plt.axis("off")

    # Generated Image
    plt.subplot(2,6,image_index + 7)

    plt.imshow(
        reconstructed_output[image_index].reshape(28,28),
        cmap='binary'
    )

    plt.axis("off")

plt.show()
```

Conclusion

Successfully implemented an unsupervised generative model using Autoencoder architecture in TensorFlow and Keras for image reconstruction.
