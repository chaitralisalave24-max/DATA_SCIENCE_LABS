Aim

To implement Deep Natural Language Processing (NLP) using TensorFlow and Keras for text classification.

Program

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import (
    Embedding,
    Bidirectional,
    LSTM,
    Dense,
    Dropout
)

# Dataset Configuration
maximum_words = 15000
sequence_length = 250

# Load IMDB Review Dataset
(train_reviews, train_targets), (test_reviews, test_targets) = imdb.load_data(
    num_words=maximum_words
)

# Apply Padding
train_reviews = pad_sequences(
    train_reviews,
    maxlen=sequence_length,
    padding='post'
)

test_reviews = pad_sequences(
    test_reviews,
    maxlen=sequence_length,
    padding='post'
)

print("Training Data Shape :", train_reviews.shape)
print("Testing Data Shape :", test_reviews.shape)

# Create NLP Sequential Model
sentiment_model = Sequential()

# Word Embedding Layer
sentiment_model.add(
    Embedding(
        input_dim=maximum_words,
        output_dim=64,
        input_length=sequence_length
    )
)

# Bidirectional LSTM Layer
sentiment_model.add(
    Bidirectional(
        LSTM(
            64,
            dropout=0.3
        )
    )
)

# Dense Hidden Layer
sentiment_model.add(Dense(
    32,
    activation='relu'
))

# Dropout Layer
sentiment_model.add(Dropout(0.4))

# Output Layer
sentiment_model.add(Dense(
    1,
    activation='sigmoid'
))

# Display Model Structure
sentiment_model.summary()

# Compile Deep Learning Model
sentiment_model.compile(
    optimizer='rmsprop',
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Train NLP Model
nlp_history = sentiment_model.fit(
    train_reviews,
    train_targets,
    epochs=4,
    batch_size=128,
    validation_split=0.25,
    verbose=1
)

# Evaluate Model
final_loss, final_accuracy = sentiment_model.evaluate(
    test_reviews,
    test_targets,
    verbose=0
)

print("\nDeep NLP Accuracy :", final_accuracy)

# Accuracy Plot
plt.figure(figsize=(7,5))

plt.plot(
    nlp_history.history['accuracy']
)

plt.plot(
    nlp_history.history['val_accuracy']
)

plt.title("NLP Model Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")

plt.legend([
    "Training Accuracy",
    "Validation Accuracy"
])

plt.show()

# Loss Plot
plt.figure(figsize=(7,5))

plt.plot(
    nlp_history.history['loss']
)

plt.plot(
    nlp_history.history['val_loss']
)

plt.title("NLP Model Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")

plt.legend([
    "Training Loss",
    "Validation Loss"
])

plt.show()
```

Conclusion

Successfully implemented Deep NLP using Embedding and LSTM layers for sentiment classification on the IMDB movie review dataset.
