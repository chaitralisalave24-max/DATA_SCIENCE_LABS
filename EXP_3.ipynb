```python
# Experiment: Neural Recommendation System using Embeddings

# Step 1: Import Libraries
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np

# Step 2: Sample User-Movie Dataset
# User IDs
users = np.array([0, 1, 2, 3, 4])

# Movie IDs
movies = np.array([0, 1, 2, 3, 4])

# Ratings given by users
ratings = np.array([5, 4, 3, 5, 2])

# Step 3: Build Recommendation Model

# User Input
user_input = keras.Input(shape=(1,))

# Movie Input
movie_input = keras.Input(shape=(1,))

# Embedding Layers
user_embedding = layers.Embedding(
    input_dim=5,
    output_dim=4
)(user_input)

movie_embedding = layers.Embedding(
    input_dim=5,
    output_dim=4
)(movie_input)

# Flatten Embeddings
user_vec = layers.Flatten()(user_embedding)
movie_vec = layers.Flatten()(movie_embedding)

# Combine Both Embeddings
concat = layers.Concatenate()([user_vec, movie_vec])

# Dense Layers
dense1 = layers.Dense(16, activation='relu')(concat)
output = layers.Dense(1)(dense1)

# Final Model
model = keras.Model(
    inputs=[user_input, movie_input],
    outputs=output
)

# Step 4: Compile Model
model.compile(
    optimizer='adam',
    loss='mse',
    metrics=['mae']
)

# Step 5: Train Model
model.fit(
    [users, movies],
    ratings,
    epochs=100,
    verbose=1
)

# Step 6: Predict Ratings
predicted_rating = model.predict([
    np.array([1]),
    np.array([3])
])

print("\nPredicted Rating for User 1 on Movie 3:")
print(predicted_rating[0][0])
```

Part B: Explicit Feedback Neural Recommendation System

```python
# Part B: Explicit Feedback Neural Recommendation System

# Step 1: Import Libraries
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np

# Step 2: Sample Dataset

# User IDs
user_ids = np.array([
    0, 0, 1, 1, 2,
    2, 3, 3, 4, 4
])

# Movie IDs
movie_ids = np.array([
    0, 1, 1, 2, 0,
    2, 1, 3, 2, 4
])

# Ratings given by users
ratings = np.array([
    5, 4, 4, 5, 3,
    4, 2, 5, 4, 3
], dtype=np.float32)

# Step 3: Define Number of Users and Movies
num_users = 5
num_movies = 5

# Step 4: Create Inputs
user_input = keras.Input(shape=(1,))
movie_input = keras.Input(shape=(1,))

# Step 5: Create Embedding Layers
user_embedding = layers.Embedding(
    input_dim=num_users,
    output_dim=8
)(user_input)

movie_embedding = layers.Embedding(
    input_dim=num_movies,
    output_dim=8
)(movie_input)

# Step 6: Flatten Embeddings
user_vector = layers.Flatten()(user_embedding)
movie_vector = layers.Flatten()(movie_embedding)

# Step 7: Concatenate Features
combined = layers.Concatenate()([
    user_vector,
    movie_vector
])

# Step 8: Dense Neural Network
dense1 = layers.Dense(64, activation='relu')(combined)
dense2 = layers.Dense(32, activation='relu')(dense1)

# Output Layer
output = layers.Dense(1)(dense2)

# Step 9: Build Model
model = keras.Model(
    inputs=[user_input, movie_input],
    outputs=output
)

# Step 10: Compile Model
model.compile(
    optimizer='adam',
    loss='mse',
    metrics=['mae']
)

# Step 11: Display Model Summary
model.summary()

# Step 12: Train Model
history = model.fit(
    [user_ids, movie_ids],
    ratings,
    epochs=100,
    verbose=1
)

# Step 13: Predict Rating

test_user = np.array([1])
test_movie = np.array([3])

predicted_rating = model.predict([
    test_user,
    test_movie
])

print("\nPredicted Rating:")
print("User ID:", test_user[0])
print("Movie ID:", test_movie[0])
print("Predicted Rating:", predicted_rating[0][0])
```

Part C: Implicit Feedback Recommendation System using Triplet Loss

```python
# Part C: Implicit Feedback Recommendation System using Triplet Loss
# ERROR-FREE VERSION FOR GOOGLE COLAB

# Step 1: Import Libraries
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np

# Step 2: Sample Dataset

# Anchor Users
anchor_users = np.array([0, 1, 2, 3])

# Positive Items (liked items)
positive_items = np.array([1, 2, 3, 4])

# Negative Items (unliked items)
negative_items = np.array([4, 0, 1, 2])

# Number of users and items
num_users = 5
num_items = 5

# Embedding dimension
embedding_dim = 8

# Step 3: Define Inputs
anchor_input = keras.Input(shape=(1,))
positive_input = keras.Input(shape=(1,))
negative_input = keras.Input(shape=(1,))

# Step 4: Embedding Layers
user_embedding = layers.Embedding(
    input_dim=num_users,
    output_dim=embedding_dim
)

item_embedding = layers.Embedding(
    input_dim=num_items,
    output_dim=embedding_dim
)

# Step 5: Create Embeddings
anchor_vec = layers.Flatten()(user_embedding(anchor_input))
positive_vec = layers.Flatten()(item_embedding(positive_input))
negative_vec = layers.Flatten()(item_embedding(negative_input))

# Step 6: Distance Calculation using Lambda Layers

# Positive Distance
positive_distance = layers.Lambda(
    lambda x: tf.reduce_sum(tf.square(x[0] - x[1]), axis=1, keepdims=True)
)([anchor_vec, positive_vec])

# Negative Distance
negative_distance = layers.Lambda(
    lambda x: tf.reduce_sum(tf.square(x[0] - x[1]), axis=1, keepdims=True)
)([anchor_vec, negative_vec])

# Step 7: Triplet Loss Calculation
margin = 1.0

triplet_loss = layers.Lambda(
    lambda x: tf.maximum(x[0] - x[1] + margin, 0.0)
)([positive_distance, negative_distance])

# Step 8: Build Model
model = keras.Model(
    inputs=[
        anchor_input,
        positive_input,
        negative_input
    ],
    outputs=triplet_loss
)

# Step 9: Compile Model
model.compile(
    optimizer='adam',
    loss=lambda y_true, y_pred: y_pred
)

# Dummy labels (required for training)
dummy_labels = np.zeros((len(anchor_users), 1))

# Step 10: Train Model
model.fit(
    [anchor_users, positive_items, negative_items],
    dummy_labels,
    epochs=100,
    verbose=1
)

# Step 11: Show Learned Embeddings

print("\nUser Embeddings:\n")
print(user_embedding.get_weights()[0])

print("\nItem Embeddings:\n")
print(item_embedding.get_weights()[0])
```
